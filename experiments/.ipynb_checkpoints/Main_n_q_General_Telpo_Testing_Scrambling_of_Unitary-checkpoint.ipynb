```python
########################## DEFINE VARIABLES FOR EXPERIMENTAL SETUP ##########################
#############################################################################################

# size of the scrambling unitary / operator
n = 4

# define two qubit gates for scrambling
two_q_gates = [CZGate()]

# define one qubit gates for scrambling
one_q_gates = [HGate()]
```

```python
########################## GENERATE CIRC STRUCTURING AUTOMATICALLY ##########################
#############################################################################################

# num of qubits in circuit = 1 q for bob + n*2
qnum = 2*n +1

# inums of upper and lower half qubits
fst_half, sec_half = fst_n_sec_half_nums(qnum)

# initial bell pair inums defining the many-body-system
bell_pairs = get_bell_pairs(qnum)

# list of all bell pairs that can be used as basemeas pair
meas_pairs = [x for x in bell_pairs[:-1]]+[[0,qnum-2]] # replace bob and neighbor with alice and bob neighbor

# pairs to apply two qubit gates to in the scrambling protocol
fst_2gate_pairs, sec_2gate_pairs = get_unitary_pairs(qnum)

# reversed order for all items but the first for gate lists and pair lists
rev_fst_2gate_pairs = [fst_2gate_pairs[0]] + [x for x in reversed(fst_2gate_pairs[1:])]
rev_sec_2gate_pairs = [sec_2gate_pairs[0]] + [x for x in reversed(sec_2gate_pairs[1:])]
rev_two_q_gates =  [two_q_gates[0]] + [x for x in reversed(two_q_gates[1:])]
rev_one_q_gates = [one_q_gates[0]] + [x for x in reversed(one_q_gates[1:])]

# alice has qubit 0 - bob has qubit with last inum 
inum_alice, inum_bob = 0, qnum-1

# define qubit pair that will be used for base measurement 
# fist pair in the list is the middle pair (furthes away from bob and alice), but any pair will do
basemeas = meas_pairs[0]

# repeat gates if selection is smaller than num of pairs in unitary
if len(two_q_gates) < len(fst_2gate_pairs):
    two_q_gates = (1+int(len(fst_2gate_pairs)/len(two_q_gates)))*two_q_gates
two_q_gates = [two_q_gates[i] for i in range(len(fst_2gate_pairs))]

if len(one_q_gates) < n*2:
    one_q_gates = (1+int(n*2/len(one_q_gates)))*one_q_gates
one_q_gates = [one_q_gates[i] for i in range(n*2)]

if len(rev_two_q_gates) < len(rev_fst_2gate_pairs):
    rev_two_q_gates = (1+int(len(rev_fst_2gate_pairs)/len(rev_two_q_gates)))*rev_two_q_gates
rev_two_q_gates = [rev_two_q_gates[i] for i in range(len(rev_fst_2gate_pairs))]

if len(rev_one_q_gates) < n*2:
    rev_one_q_gates = (1+int(n*2/len(rev_one_q_gates)))*rev_one_q_gates
rev_one_q_gates = [rev_one_q_gates[i] for i in range(n*2)]
```

```python
########################## GENERATE CIRCUIT FOR EXPERIMENT ##################################
#############################################################################################

# initialize circ 
qr, cr = QuantumRegister(qnum), ClassicalRegister(qnum)
circ = QuantumCircuit(qr, cr)

# initialize alice qubit in |1> state
circ.x(inum_alice)

# initialize bell pairs
entangle_bell_pairs(circ, bell_pairs)

# apply two qubit gates for pairs in fist half of circ
for i in range(len(fst_2gate_pairs)):
    circ.append(two_q_gates[i],fst_2gate_pairs[i])
    
# apply two qubit gates for pairs in second half of circ
for i in range(len(sec_2gate_pairs)):
    circ.append(two_q_gates[i],sec_2gate_pairs[i])

# apply single qubit gates for all but bob qubit
for i in range(2*n):
    circ.append(one_q_gates[i],[i])

# apply two qubit gates for reversed pairs in fist half of circ
for i in range(len(rev_fst_2gate_pairs)):
    circ.append(rev_two_q_gates[i],rev_fst_2gate_pairs[i])
    
# apply two qubit gates for reversed pairs in second half of circ
for i in range(len(rev_sec_2gate_pairs)):
    circ.append(rev_two_q_gates[i],rev_sec_2gate_pairs[i])

circ.barrier()
# disentangle base meas bell pair
disentangle_bell_pair(circ, basemeas)

# apply decoding from measurement
apply_bob_gates(circ,inum_bob,basemeas)

# measure bob qubit
circ.measure(inum_bob,inum_bob)

circ.draw('mpl')
```

```python
# choose simulator 

be = qasm_simulator

# run on simulator 

pct = 100
shots = 100*pct
t_qc = transpile(circ, be)
qobj = assemble(t_qc, shots=shots)
result = be.run(qobj).result()

# see fidelity

bob_1 = sum([v for (k,v) in result.get_counts().items() if k[0]== "1"])
bob_0 = sum([v for (k,v) in result.get_counts().items() if k[0]== "0"])
print("\n Fidelity Bob Qubit:   \t {}%".format(bob_1/pct))
fid_dict = {"1":bob_1,"0":bob_0} # dict for hist plot
plot_histogram(fid_dict)
```
