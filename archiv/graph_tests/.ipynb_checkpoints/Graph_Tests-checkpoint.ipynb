# Testing Graphs and Circs via Entangler Map

https://qiskit.org/documentation/locale/de_DE/tutorials/algorithms/06_qaoa.html?highlight=entanglement

```python
from qiskit.utils import get_entangler_map
import networkx as nx
from qiskit.circuit.library import GraphState
```

```python
# set num of qubits and offset
num_qubits = 5
offset = 0

# store pairings for all  all three types of entanglment
linear_e = [tuple(x) for x in get_entangler_map('linear', num_qubits,offset )]
sca_e = [tuple(x) for x in get_entangler_map('sca',num_qubits,offset )]
full_e = [tuple(x) for x in get_entangler_map('full', num_qubits,offset)]

# see lists made via all three types of entanglment
print("full:\t",full_e,'\nlinear:\t',linear_e,'\nsca:\t',sca_e)
```

## Linear Entanglement

```python
# make graph and circ 
g = nx.from_edgelist(linear_e) # get graph
matrix = nx.to_numpy_matrix(g) # get matrix for GraphState circ
graph_circ =GraphState(matrix) # get circ

# see circ depth
print("\nCirc Depth:",graph_circ.depth())

# see graph and circ
nx.draw(g)
graph_circ.draw('mpl')
```

## Circular Entanglement

```python
# make graph and circ 
g = nx.from_edgelist(sca_e) # get graph
matrix = nx.to_numpy_matrix(g) # get matrix for GraphState circ
graph_circ =GraphState(matrix) # get circ

# see circ depth
print("\nCirc Depth:",graph_circ.depth())

# see graph and circ
nx.draw(g)
graph_circ.draw('mpl')
```

## Full Entanglement

```python
# make graph and circ 
g = nx.from_edgelist(full_e) # get graph
matrix = nx.to_numpy_matrix(g) # get matrix for GraphState circ
graph_circ =GraphState(matrix) # get circ

# see circ depth
print("\nCirc Depth:",graph_circ.depth())

# see graph and circ
nx.draw(g)
graph_circ.draw('mpl')
```
